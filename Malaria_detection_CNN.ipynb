## Reading the all images and also doing encode

```
labels = []
data = []
for each in os.listdir():
  for img in os.listdir(each):
    try:
      img_path = '/content/drive/My Drive/TASKS_ML_NETZWERK_ACADAMEY/DL/Malaria_detection/cell_images/'+each+'/'+img
      # print(img_path)
      img_read = cv2.imread(img_path,0)
      # cv2_imshow(img_read)
      img_resize = cv2.resize(img_read,(100,100))
      data.append(img_resize)
      if each == 'Parasitized':
        labels.append(1)
      else:
        labels.append(0)
    except AssertionError:
      pass
    except:
      pass
```

## Converting data into arrays

```
x = np.array(data)
y = np.array(labels)
x.shape
```

## Changing shape into 100,100,1

```
x = x.reshape(x.shape[0],x.shape[1],x.shape[2],1)
x.shape
```

```
# y = np_utils.to_categorical(y,2)
```

## Splitting the data into train and test data

```
x_train,x_test,y_train,y_test = train_test_split(x,y,train_size=0.7,random_state=1)
x_train.shape,x_test.shape,y_train.shape,y_test.shape
```

#Creating model

```
model = Sequential()#sequential model
```

```
## 3convonual layer and 1 max pooling 
model.add(Conv2D(filters=64,kernel_size=(3,3),activation='relu',input_shape=(100,100,1)))
model.add(Conv2D(filters=62,kernel_size=(3,3),activation='relu'))
model.add(Conv2D(filters=32,kernel_size=(3,3),activation='relu'))
model.add(MaxPooling2D())
## 2 convonual layer and 1 max pooling
model.add(Conv2D(filters=32,kernel_size=(3,3),activation='relu'))
model.add(Conv2D(filters=32,kernel_size=(3,3),activation='relu'))
model.add(MaxPooling2D())
## Flattening the cnn output
model.add(Flatten())
## 2Hidden layers
model.add(Dense(units=64,activation='relu'))
model.add(Dense(units=32,activation='relu'))
##Dropout 
model.add(Dropout(rate=0.4))
##Output layer
model.add(Dense(1,activation='sigmoid'))
```

Summary

```
model.summary()
```

##Compiling our model

```
model.compile(optimizer='adam',loss='binary_crossentropy',metrics=['accuracy'])
```

```
model.fit(x=x_train,y=y_train,batch_size=20,epochs=15,validation_data=(x_test,y_test))
```

```
model.evaluate(x_test,y_test)
```

```
model.save('Malaria_detection.h5')
```
